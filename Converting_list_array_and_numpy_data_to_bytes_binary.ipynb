## struct.pack(format, v1, v2, ...)
: format 에 따라 압축된 v1 , v2 , … 값을 포함하는 바이트열 객체 반환

```python
from struct import *
```

```python
pack('h', 1)
```

```python
pack('h', 2)
```

```python
pack('i', 3)
```

```python
pack('hhi', 1, 2, 3)
```

## struct.unpack(format, buffer)
: 문자열 형식에 따라 버퍼에서 압축 풀기

```python
unpack('hhi', b'\x01\x00\x02\x00\x03\x00\x00\x00')
```

## struct.calcsize(format)
: 문자열 format에 해당하는 구조체의 크기(pack(format, v1, v2, ...)에 의해 생성된 바이트열 객체의 크기) 반환

```python
calcsize('hhl') # 2 + 2 + 4
```

```python
calcsize('iih') # 4 + 4 + 2
```
